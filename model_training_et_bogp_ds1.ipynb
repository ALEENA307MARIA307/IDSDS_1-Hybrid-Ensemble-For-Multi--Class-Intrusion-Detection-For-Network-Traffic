```python
import pandas as pd
df_train=pd.read_csv("train.csv")
df_test=pd.read_csv("test_f.csv")
```

```python
# Separate features and target
X_train = df_train.drop(columns=["label"])
y_train = df_train["label"]
# Separate features and target
X_test = df_test.drop(columns=["label"])
y_test = df_test["label"]
```

```python
from sklearn.ensemble import ExtraTreesClassifier
import numpy as np
```

```python
import joblib
```

```python
best_et_BOGP=joblib.load("best_et_BOGP.pkl")
```

```python
clf_et=ExtraTreesClassifier(**best_et_BOGP,random_state=42)
```

```python
clf_et.fit(X_train,y_train)
```

```python
y_pred = clf_et.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import accuracy_score, balanced_accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, classification_report
```

```python
# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
balanced_acc = balanced_accuracy_score(y_test, y_pred)
f1_macro = f1_score(y_test, y_pred, average='macro')
f1_weighted = f1_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
cm = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy:.4f}")
print(f"Balanced Accuracy: {balanced_acc:.4f}")
print(f"F1 Score (Macro): {f1_macro:.4f}")
print(f"F1 Score (Weighted): {f1_weighted:.4f}")
print(f"Precision (Weighted): {precision:.4f}")
print(f"Recall (Weighted): {recall:.4f}")
print("\nConfusion Matrix:\n", cm)
print("\nClassification Report:\n", report)
```
