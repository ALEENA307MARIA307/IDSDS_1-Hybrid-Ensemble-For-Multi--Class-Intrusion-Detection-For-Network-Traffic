```python
import pandas as pd
```

```python
df=pd.read_csv("train_f.csv")
```

```python
X = df.drop('label', axis=1)
y = df['label']
```

```python

df['label'].value_counts()
```

```python
import numpy as np
```

```python
#Apply SMOTE to training data
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_r, y_r = smote.fit_resample(X, y)
```

```python
y_r.value_counts()
```

```python
# Combine into one DataFrame
df_resampled = pd.concat([pd.DataFrame(X_r, columns=X.columns),pd.DataFrame(y_r, columns=['label'])],axis=1)

df_shuffled = df_resampled.sample(frac=1, random_state=42).reset_index(drop=True)

# Save to CSV
df_shuffled.to_csv("train.csv", index=False)
```
