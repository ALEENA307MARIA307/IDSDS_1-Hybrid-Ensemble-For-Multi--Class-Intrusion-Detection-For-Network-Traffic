```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
import joblib
```

```python
best_dt_ids=joblib.load("best_dt_BOGP.pkl")
best_et_ids=joblib.load("best_et_BOGP.pkl")
best_rf_ids=joblib.load("best_rf_BOGP.pkl")
```

```python
dt=DecisionTreeClassifier(**best_dt_ids,random_state=42)
et=ExtraTreesClassifier(**best_et_ids,random_state=42)
rf=RandomForestClassifier(**best_rf_ids,random_state=42)
```

```python
df_train1=pd.read_csv("train.csv")
#df_train2=pd.read_csv("trainv.csv")
df_test=pd.read_csv("test_f.csv")
```

```python
# Separate features and target
X_train1 = df_train1.drop(columns=["label"])
y_train1 = df_train1["label"]

#X_train2 = df_train2.drop(columns=["label"])
#y_train2 = df_train2["label"]

X_test = df_test.drop(columns=["label"])
y_test = df_test["label"]
```

```python
for model in [dt,et,rf]:
    model.fit(X_train1,y_train1)
```

```python
joblib.dump(dt, "dt_ids.pkl")
joblib.dump(et, "et_ids.pkl")
joblib.dump(rf, "rf_ids.pkl")
```

```python
# Get base model probabilities on meta training set
probs_meta = [model.predict_proba(X_train1) for model in [dt, et, rf]]
```

```python
prob_dfs = []
for i, probs in enumerate(probs_meta, start=1):
    # probs.shape = (n_samples, n_classes)
    n_classes = probs.shape[1]
    col_names = [f'model{i}_class{j}' for j in range(n_classes)]
    prob_dfs.append(pd.DataFrame(probs, columns=col_names))
```

```python
meta_features_df = pd.concat(prob_dfs, axis=1)
```

```python
meta_features_df['label'] = y_train1
```

```python
meta_features_df.to_csv('meta_train.csv', index=False)
```
