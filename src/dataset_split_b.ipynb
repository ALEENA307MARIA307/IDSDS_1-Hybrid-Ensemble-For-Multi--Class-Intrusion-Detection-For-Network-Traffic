```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("pretrain_dataset.csv")
```

```python
X = df.drop("label", axis=1)
y = df["label"]
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X, 
    y, 
    test_size=0.2, 
    stratify=y, 
    random_state=42
)
```

```python
train_df = pd.concat([X_train, y_train], axis=1)
val_df = pd.concat([X_val, y_val], axis=1)
```

```python
train_df.head()
```

```python
train_df.to_csv("traint.csv", index=False)
val_df.to_csv("trainv.csv", index=False)
```
