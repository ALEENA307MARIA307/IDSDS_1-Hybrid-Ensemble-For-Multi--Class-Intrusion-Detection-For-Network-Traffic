```python
import numpy as np
import pandas as pd

def correlation_filter(X, threshold=0.9):

    corr_matrix = X.corr().abs()

    # Track columns to drop
    dropped_features = set()
    
    # Iterate through correlation matrix
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in dropped_features):
                # Drop the current feature
                col_to_drop = corr_matrix.columns[i]
                dropped_features.add(col_to_drop)
    
    # Create new dataframe with remaining features
    X_filtered = X.drop(columns=list(dropped_features), errors='ignore')

    
    print(f"Correlation threshold: {threshold}")
    print(f"Dropped {len(dropped_features)} highly correlated features.")
    print(f"Remaining features: {X_filtered.shape[1]}")
    print(f" New feature set:\n{list(X_filtered.columns)}")

    return X_filtered, list(dropped_features)


X_cf, dropped = correlation_filter(X, threshold=0.9)
```

```python
#random forest feature importance
rf = RandomForestClassifier(random_state=42, n_jobs=-1)
rf.fit(X_cf, y)

importances = pd.Series(rf.feature_importances_, index=X_cf.columns).sort_values(ascending=False)
print("Top 10 features by RF importance:")
print(importances.head(10))
```

```python
sf=X_cf.columns
```

```python
if IMPORTANCE_THRESHOLD is not None:
    selected_features = importances[importances >= IMPORTANCE_THRESHOLD].index.tolist()
    print(f"Selected {len(selected_features)} features by importance threshold >= {IMPORTANCE_THRESHOLD}")
elif TOP_K is not None:
    selected_features = importances.head(TOP_K).index.tolist()
    print(f"Selected top {len(selected_features)} features by RF importance (TOP_K={TOP_K})")
else:
    # fallback - keep all (not recommended)
    selected_features = sf.tolist()
```
